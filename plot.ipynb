```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = np.loadtxt('scenario_320K.txt').T
data2 = np.loadtxt('scenario_280K.txt').T
```

```python
fig,ax=plt.subplots(figsize=(8, 6),dpi=600,constrained_layout=False)
ax.set_title('First scenario (constant albedo)')
ax.plot(data1[0], data1[1], label='$T_0=320$')
ax.plot(data2[0], data2[1], label='$T_0=280$')
ax.set_xlabel('t')
ax.set_ylabel('T')
ax.grid()
ax.legend()
```

```python
data3 = np.loadtxt('scenario2_250K.txt').T
data4 = np.loadtxt('scenario2_270K.txt').T
```

```python
fig,ax=plt.subplots(figsize=(8, 6),dpi=600,constrained_layout=False)
ax.set_title('Second scenario (temperature-dependent albedo)')
ax.plot(data3[0], data3[1], label='$T_0=250$')
ax.plot(data4[0], data4[1], label='$T_0=270$')
ax.set_xlabel('t')
ax.set_ylabel('T')
ax.grid()
ax.legend()
```
